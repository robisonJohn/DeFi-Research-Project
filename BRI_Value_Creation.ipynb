Our objective for the BRI data is simple:
For each cryptocurrency, we need to measure the following key metrics over time:
1. Value Creation
2. Market Capitalization
3. Transaction Volume
4. Velocity
We will start with Value Creation for Ethereum. Ideally, we will then generate value creation for each of our tokens in question. We will then proceed to generate each additional metric for our tokens in question. Ultimately, this will give us a strong visual understanding of each token's relative dominance over time.

```python
ethUrl = 'https://api.lunarcrush.com/v2?data=assets&key=sou5nb5qo32mso99umfel&symbol=ETH&interval=day&data_points=400'
ethResponse = requests.get(ethUrl)
```

```python
json_eth = ethResponse.json()
json_eth['data'][0]['timeSeries'][0]
```

```python
ethTimeSeries = json_eth['data'][0]['timeSeries']
```

```python
ethPriceData = [item['close'] for item in ethTimeSeries]
print(ethPriceData)
```

```python
import matplotlib.pyplot as plt

indices = [i for i in range(400)]

plt.scatter(ethPriceData, indices)
plt.show()
```

```python
import numpy as np

x = np.array(indices)
y = np.array(ethPriceData)

m, b = np.polyfit(x,y,1)

plt.plot(x, y, 'o')
plt.plot(x, m*x + b)
```

In general, to model each time series, we will need to do the following:
1. Pull api data for the asset in question
2. Pull the time series data from the json format
3. Graph the time series data as a function of time

```python
def retrieveData(symbol, data_points, metric, name):
  coinUrl = f'https://api.lunarcrush.com/v2?data=assets&key=sou5nb5qo32mso99umfel&symbol={symbol}&interval=day&data_points={data_points}'
  coinResponse = requests.get(coinUrl)
  json_coin = coinResponse.json()
  coinTimeSeries = json_coin['data'][0]['timeSeries']
  coinPriceData = [item[metric] for item in coinTimeSeries]
  indices = [i for i in range(data_points)]
  plt.figure(figsize=(8,5))
  if metric == 'close':
    plt.title(f'{name} price data at market close over prior {data_points} days')
    plt.xlabel('Days')
    plt.ylabel(f'Price of {name}')
    plt.scatter(indices, coinPriceData)
  else:
    if metric == 'market_cap':
      # put stuff here
      plt.title(f'{name} market capitalization over prior {data_points} days')
      plt.xlabel('Days')
      plt.ylabel(f'Market Capitalization of {name}')
      plt.scatter(indices, coinPriceData)
    else:
      if metric == 'volume':
        # put stuff here
        plt.title(f'{name} transaction volume over prior {data_points} days')
        plt.xlabel('Days')
        plt.ylabel(f'Transaction Volume of {name}')
        plt.scatter(indices, coinPriceData)
  return plt.show()
  

  
```
